```python
print ( "----------------------- Registro Civil Totem Autoservicio-------------------------")
personas=[]


def validar_rut(rut):
    rut = rut.replace("-", "").replace(".", "")
    if len(rut) != 9:
        return False
    if not rut[:-1].isdigit():
        return False

    digito_verificador = rut[-1].upper()
    rut_numeros = list(map(int, rut[:-1]))

    multiplicador = 2
    suma = 0
    for i in range(len(rut_numeros)-1, -1, -1):
        suma += rut_numeros[i] * multiplicador
        multiplicador = (multiplicador + 1) % 8 or 2

    digito_calculado = 11 - (suma % 11)
    digito_calculado = str(digito_calculado) if digito_calculado != 10 else "K"

    return digito_calculado == digito_verificador


def guardar():
  rut = input("Ingrese un RUT: ")
  if validar_rut(rut):
    print("RUT válido")
    nombre = input("Nombre: ")
    if len(nombre) < 2:
     print("El nombre debe tener mínimo 2 caracteres")
     return
    edad = int(input("Edad: "))
    if edad < 0:
     print("La edad debe ser mayor a 0")
     return

    Nacionalidad = input("Nacionalidad: ")
    ciudad_nacimiento = input("Ciudad de nacimiento: ")

    persona = {
        "rut": rut,
        "nombre": nombre,
        "edad": edad,
        "Nacionalidad": Nacionalidad,
        "ciudad_nacimiento": ciudad_nacimiento
        }
    return persona
  else:
    print("RUT inválido")
    return
def buscar(personas):
    rut = input("Ingrese el RUT a buscar: ")
    for persona in personas:
        if persona["rut"] == rut:
            return persona
    return None
def eliminar(personas):
    rut = input("Ingrese el RUT: ")
    for persona in personas:
        if persona["rut"] == rut:
            personas.remove(persona)
            print("Registro Eliminado ")
            return
    print("Rut no registrado")

while True:

        print("----- MENÚ --------------------")
        print("-----1.-Guardar usuario--------")
        print("-----2.-Buscar-----------------")
        print("-----3.-Imprimir certificados--")
        print("-----4.-Eliminar---------------")
        print("-----5.-Salir------------------")
        try:
          op=int(input("Ingrese opcion: "))
          while op< 1 or op> 5 :
           print("Ingresar Opcion Valida ")
           op=int(input("Ingrese opcion: "))
          if op==1:
             persona=guardar()
             if persona:
              personas.append(persona)
              print("Registro Guardado")
              print(personas)
          elif op==2:
           while True:
            try:
              persona = buscar(personas)
              if persona:
                print("Información de la persona:")
                print("RUT: ", persona["rut"])
                print("Nombre: ", persona["nombre"])
                print("Edad: ", persona["edad"])
                print("Nacionalidad: ", persona["Nacionalidad"])
                print("Ciudad de nacimiento: ", persona["ciudad_nacimiento"])

                if persona["Nacionalidad"].lower() == "chileno":
                    print("La persona es de Nacionalidad Chilena")
                else:
                    print("La persona no es chilena")
              break
            except ValueError:
                print("Persona no encontrada ")



          elif op==3:
             rut = input("Ingrese el RUT : ")
             for persona in personas:
              if persona["rut"] == rut:
                 print("Certificado de Nacimiento")
                 print("Nombre: ", persona["nombre"])
                 print("RUT: ", persona["rut"])
                 print("Nacionalidad: ", persona["Nacionalidad"])
                 print("Ciudad de nacimiento: ", persona["ciudad_nacimiento"])
              else:
                print("Persona no encontrada")

          elif op== 4:
            eliminar(personas)
          elif op==5:
           print("Cerrando sistema")
           break

        except ValueError:
           print("---ingresar solo numeros indicados en el menu-----")






```
